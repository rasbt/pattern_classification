## Deriving the decision boundary
[<a href="#sections">back to top</a>] <br>

### Bayes' Rule:


$ P(\omega_j|x) = \frac{p(x|\omega_j) * P(\omega_j)}{p(x)}$ 


### Risk Functions:

$ R(\alpha_1|\;\vec{x}) = \lambda_{11}P(\omega_1|\;\vec{x}) + \lambda_{12}P(\omega_2|\;\vec{x}) $

$ R(\alpha_2|\;\vec{x}) = \lambda_{21}P(\omega_1|\;\vec{x}) + \lambda_{22}P(\omega_2|\;\vec{x}) $

<br>
with 1-0 loss function:

$ R(\alpha_1|\; \vec{x}) = P(\omega_2|\;\vec{x}) = 1 -  P(\omega_1|\;\vec{x})  $

$ R(\alpha_2|\;\vec{x}) = P(\omega_1|\;\vec{x}) = 1 -  P(\omega_2|\;\vec{x}) $

### Discriminant Functions:
$ g_1(\vec{x}) = P(\omega_1 | \; \vec{x}), \quad  g_2(\vec{x}) = P(\omega_2 | \; \vec{x})$

$ \Rightarrow g_1(\vec{x}) = P(\vec{x}|\;\omega_1) \;\cdot\; P(\omega_1) \quad | \; ln \\
\quad g_2(\vec{x}) = P(\vec{x}|\;\omega_2) \;\cdot\; P(\omega_2) \quad | \; ln $

<br>
We can drop the prior probabilities (since we have eual priors in this case): 

$ \Rightarrow g_1(\vec{x}) = ln(P(\vec{x}|\;\omega_1))  \\
\quad g_2(\vec{x}) = ln(P(\vec{x}|\;\omega_2))  $

$ \Rightarrow g_1(\vec{x}) = \vec{x}^{\,t} - \frac{1}{2} \Sigma_1^{-1} \vec{x} + \bigg( \Sigma_1^{-1} \vec{\mu}_{\,1}\bigg)^t + \bigg( -\frac{1}{2} \vec{\mu}_{\,1}^{\,t} \Sigma_{1}^{-1} \vec{\mu}_{\,1} -\frac{1}{2} ln(|\Sigma_1|)\bigg)   \\
\quad g_2(\vec{x}) = \vec{x}^{\,t} - \frac{1}{2} \Sigma_2^{-1} \vec{x} + \bigg( \Sigma_2^{-1} \vec{\mu}_{\,2}\bigg)^t + \bigg( -\frac{1}{2} \vec{\mu}_{\,2}^{\,t} \Sigma_{2}^{-1} \vec{\mu}_{\,2} -\frac{1}{2} ln(|\Sigma_2|)\bigg)  $

<hr>
**Let:**

$\vec{W}_{\,i} = - \frac{1}{2} \Sigma_i^{-1}\\
\vec{w}{\,i} =  \bigg( \Sigma_i^{-1} \vec{\mu}_{\,i} \bigg)^t\\
\omega_{i0} = \bigg( -\frac{1}{2} \vec{\mu}_{\,i}^{\,t} \Sigma_{i}^{-1} \vec{\mu}_{\,i} -\frac{1}{2} ln(|\Sigma_i|)\bigg)$
<hr>

$ \vec{W}_{\,1} = \bigg[ 
\begin{array}{cc}
(1/4) & 0\\
0 & (1/4) \\
\end{array} \bigg] $

$ \vec{w}_{\,1} = \bigg[ 
\begin{array}{cc}
(1/4) & 0\\
0 & (1/4) \\
\end{array} \bigg] \cdot \bigg[ 
\begin{array}{c}
0 \\
0 \\
\end{array} \bigg] = \bigg[ 
\begin{array}{c}
0 \\
0 \\
\end{array} \bigg]$ 

$ \omega_{10} = -\frac{1}{2} [0 \quad 0 ] \bigg[ 
\begin{array}{cc}
(1/4) & 0\\
0 & (1/4) \\
\end{array} \bigg] \cdot \bigg[ 
\begin{array}{c}
0 \\
0 \\
\end{array} \bigg] 
- ln(2) = -ln(2)$

$ \vec{W}_{\,2} = \bigg[ 
\begin{array}{cc}
(-1/2) & 0\\
0 & (-1/2) \\
\end{array} \bigg] $

$ \vec{w}_{\,2} = \bigg[ 
\begin{array}{c}
1 \\
2 \\
\end{array} \bigg]$ 

$ \omega_{20} = -\frac{1}{2} [1 \quad 2 ] \bigg[ 
\begin{array}{cc}
(1/4) & 0\\
0 & (1/4) \\
\end{array} \bigg] \cdot \bigg[ 
\begin{array}{c}
1 \\
2 \\
\end{array} \bigg] 
- \frac{1}{2}ln(1) = -2.5$

$ \Rightarrow g_1(\vec{x}) = \vec{x}^{\,t} \bigg[ 
\begin{array}{cc}
(1/4) & 0\\
0 & (1/4) \\
\end{array} \bigg] 
\vec{x} + \bigg[
\begin{array}{c}
0 \\
0 \\
\end{array} \bigg]^t - ln(2) = \vec{x}^{\,t} - \frac{1}{4} \vec{x} - ln(2)$

$ \Rightarrow g_2(\vec{x}) = \vec{x}^{\,t} \bigg[ 
\begin{array}{cc}
(-1/2) & 0\\
0 & (-1/2) \\
\end{array} \bigg] 
\vec{x} + \bigg[
\begin{array}{c}
1 \\
2 \\
\end{array} \bigg]^t \vec{x} -2.5 = \vec{x}^{\,t} - \frac{1}{2} \vec{x} + [1 \quad 2] \; \vec{x} - 2.5$

### Decision Boundary

$g_1(\vec{x}) = g_2(\vec{x}) $ 

$ \Rightarrow \vec{x}^{\,t} - \frac{1}{4} \vec{x} - ln(2) = \vec{x}^{\,t} - \frac{1}{2} \vec{x} + [1 \quad 2] \; \vec{x} - 2.5 \quad \bigg| \; \cdot \; 4$

$ \Rightarrow \vec{x}^{\,t} -  \vec{x} - 4ln(2) = \vec{x}^{\,t} - 2 \vec{x}+ 4 \;( [1 \quad 2] \; \vec{x}) - 10 $

$ \Rightarrow [x_1 \quad x_2] \cdot \bigg(-\bigg[ \begin{array}{c}
x_1 \\
x_2 \\
\end{array}\bigg] \bigg) -4ln(2) = [x_1 \quad x_2] \cdot \bigg(-\bigg[ \begin{array}{c}
2x_1 \\
2x_2 \\
\end{array}\bigg] \bigg) + [4 \quad 8] \cdot 
\bigg[ \begin{array}{c}
x_1 \\
x_2 \\
\end{array}\bigg] -10$

$ \Rightarrow -x_1^2 - x_2^2 - 4ln(2) = -2x_1^2 - 2x_2^2 + 4x_1 + 8x_2 -10 $

$ \Rightarrow x_1^2 + x_2^2 -4x_1 - 8x_2 - 4ln(2) + 10 = 0 $

## Calculating the empirical error rate

[<a href="#sections">back to top</a>] <br>

```python
def decision_rule(x_vec):
    """ Returns value for the decision rule of 2-d row vectors """
    x_1 = x_vec[0]
    x_2 = x_vec[1]
    return x_1**2 + x_2**2 -4*x_1 - 8*x_2 - 4*np.log(2) + 10

w1_as_w2, w2_as_w1 = 0, 0

for x in x1_samples:
    if decision_rule(x) < 0:
        w1_as_w2 += 1
for x in x2_samples:
    if decision_rule(x) > 0:
        w2_as_w1 += 1

emp_err = (w1_as_w2 + w2_as_w1) / float(len(x1_samples) + len(x2_samples))
    
print('Empirical Error: {}%'.format(emp_err * 100))
```
